```r
#30% chance of winning
w0 <- 100
p <- 0.3

T <- 10000
time_periods <- T+1
w <- rep(0, T+1)
w[1] <- w0

X <- rep(0,T)
i <- 1

while(T>0){
    x <- runif(1)
    
    #print(sprintf("you started with %s euros.", w))

    if(x>p){
        w[i+1] <- w[i]-1
        X[i] <- 0
        #print("you lose")
    } else {
        w[i+1] <- w[i]+1
        X[i] <- 1
        #print("you win")
    }
    #print(sprintf("Now you've got %s euros.", w[i+1]))
    
    T <- T-1
    i <- i+1
}
plot(1:time_periods, w, type='l')
```
